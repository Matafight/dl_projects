```python
import tensorflow as tf
import os
```

```python
data_dir = '../inputs/verification_code_imgs/'
image_names = []
for root, sub_folder, file_list in os.walk(data_dir):
    image_names += [os.path.join(root, file_path) for file_path in file_list]
```

```python
image_names
```

```python
def get_label(str_label):
    """
    Convert the str_label to 10 binary code, 385 to 0001010010
    """
    result = [0]*2
    for i in str_label:
        result[int(i)] = 1
    return result
```

```python
import random
random.shuffle(image_names)
labels = [get_label(file_name.split('/')[-1].split('.')[0].split('_')[-1]) for file_name in image_names]
```

```python
labels
```

```python
image_ts = tf.convert_to_tensor(image_names)
```

```python
sess = tf.Session()
sess.run(image_ts)
```
